```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
def super_gauss(x, y, a, N):
    return np.exp(-((x**2 + y**2) / a**2) ** N)

def super_gauss_1d(x, a, N):
    return np.exp(-((x**2) / a**2) ** N)
```

```python
a, N, M = 0.5, 4, 512
```

```python
x, y = np.mgrid[-1:1:1j*M, -1:1:1j*M]
```

```python
plt.imshow(super_gauss(x, y, a, N))
```

```python
plt.imshow(np.diff(super_gauss(x, y, a, N), n = 1, axis = 0))
plt.colorbar()
```

```python
x_1d = np.linspace(-1,1,512)
plt.plot(x_1d, super_gauss_1d(x_1d, a, N))
plt.title('одномерный супергаусс')
```

```python
plt.plot(x_1d[:-1], np.diff(super_gauss_1d(x_1d, a, N) * np.where(np.abs(x_1d) < a/2, x_1d, np.zeros(x_1d.shape))))
plt.title('1-я разностная производная')
```

```python
def gauss_diff(x, a, N):
    return - (2 * N * np.exp(-(x/a) ** (2 * N)) * (x / a) ** (2 * N - 1)) / a
```

```python
plt.plot(x_1d, gauss_diff(x_1d,a,N) * np.where(np.abs(x_1d) < a/2, x_1d, np.zeros(x_1d.shape)))
plt.title('1-я производная')
```
